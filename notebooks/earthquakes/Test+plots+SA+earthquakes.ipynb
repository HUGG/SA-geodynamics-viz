# TEST PLOTS

#### First import modules:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Import additional modules:**

```python
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from matplotlib import colors
from matplotlib import cm
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.patches import Rectangle
```

#### Set path to data:

```python
fp = "/Users/Nellimetiainen/Desktop/SA-earthquakes.csv"
```

#### Set dataframe and datetime index:

```python
dataFrame = pd.read_csv(fp, parse_dates=True, index_col='time', sep=',')
```

## Plotting data

#### Make a simple plot showing coordinate points:

```python
# Set figure size:
plt.rcParams['figure.figsize'] = [6,12]

# Plot data. Assign x and y, mark as red dots:
plt.plot(dataFrame["longitude"], dataFrame["latitude"], "ro")

# Assign values for ticks between certain limits, assign number of ticks:
plt.xticks(np.linspace(-90, -55, 4))
plt.yticks(np.linspace(10, -60, 7))

# Add titles for the whole picture and x & y, and choose colors:
plt.title ('All earthquakes in South America 01.01.1960 - 22.05.2018', color='b')
plt.xlabel('Longitude', color='b')
plt.ylabel('Latitude', color='b')

# Display image:
plt.show()
```

**Scatter plot according to magnitude** 

```python
# Set figure size:
plt.rcParams['figure.figsize'] = [12,16]

# Plot data as a scatterplot. Assign x & y, assign size & color:
plt.scatter(dataFrame["longitude"], dataFrame["latitude"], s=dataFrame["mag"]*20, c=dataFrame["mag"], cmap='magma_r')

# Assign values for ticks between certain limits, assign number of ticks:
plt.xticks(np.linspace(-90, -55, 4))
plt.yticks(np.linspace(10, -60, 7))

# Create bounds for colorbar (min and max values for magnitude, and the number of separate values displayed):
bounds = np.linspace(4, 10, 7)
# Create colorbar:
cbar = plt.colorbar(boundaries=bounds)
# Add colorbar label:
cbar.set_label('Magnitude', color='b')

# Add titles for the whole picture and x & y, and choose colors:
plt.title ('Earthquake magnitudes 01.01.1960 - 22.05.2018', color='b')
plt.xlabel('Longitude', color='b')
plt.ylabel('Latitude', color='b')

# Display image:
plt.show()
```

**Scatter plot according to depth**

```python
# Set figure size:
plt.rcParams['figure.figsize'] = [12,16]

# Plot data as a scatterplot. Assign x & y, assign size & color:
plt.scatter(dataFrame["longitude"], dataFrame["latitude"], s=dataFrame["depth"]*2, c=dataFrame["depth"], cmap='viridis_r')

# Assign values for ticks between certain limits, assign number of ticks:
plt.xticks(np.linspace(-90, -55, 4))
plt.yticks(np.linspace(10, -60, 7))

# Create bounds for colorbar (min and max values for magnitude, and the number of separate values displayed):
bounds = np.linspace (0, 700, 8)
# Create colorbar:
cbar = plt.colorbar(boundaries=bounds)
# Add colorbar label:
cbar.set_label('Depths (km)', color='b')

# Add titles for the whole picture and x & y, and choose colors:
plt.title ('Earthquake depths 01.01.1960 - 22.05.2018', color='b')
plt.xlabel('Longitude', color='b')
plt.ylabel('Latitude', color='b')

# Display image:
plt.show()
```
