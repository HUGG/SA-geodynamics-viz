# CROSS-SECTION VERSION SA SUBDUCTION ZONE EARTHQUAKES

Creating a cross section view of earthquakes occurring in the South American subduction zone, in selected latitudes. This helps visualize the descending slab. The South American subduction zone is also interesting from the point that also flat subduction is taking place.

#### First import modules:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Import additional modules:**

```python
from mpl_toolkits.basemap import Basemap
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from matplotlib import colors
from matplotlib import cm
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.patches import Rectangle
```

#### Set path to data:

```python
fp = "/Users/Nellimetiainen/Desktop/SA-earthquakes.csv"
```

#### Set dataframe and datetime index:

```python
dataFrame = pd.read_csv(fp, parse_dates=True, index_col='time', sep=',')
```

## New columns for selected flat-slab latitudes

```python
bucaramanga = dataFrame.loc[(dataFrame['latitude'] >= 5.0000) & (dataFrame['latitude'] < 9.0)]
pfss = dataFrame.loc[(dataFrame['latitude'] <= -2.0000) & (dataFrame['latitude'] >= -14.9999)]
pampean = dataFrame.loc[(dataFrame['latitude'] <= -27.0000) & (dataFrame['latitude'] >= -32.9999)]
```

### Bucaramanga

```python
# Assign a value for the trench longitude as reference point for counting the distance (we used Google maps):
trench_longitude_buc = -78
# Calculate the distance of the trench in degrees: 
bucaramanga['trench_dist_deg_buc'] = bucaramanga['longitude'] - trench_longitude_buc
# Leave out any earthquakes located to the west of the trench:
bucaramanga = bucaramanga.drop(bucaramanga[bucaramanga['trench_dist_deg_buc'] < 0.0].index)
# Calculate the distance (km) between the trench and earthquakes:
bucaramanga['trench_dist_km_buc'] = bucaramanga['trench_dist_deg_buc'] * 111.0 * np.cos(bucaramanga['latitude']*(np.pi/180.0))

# Set figure size:
plt.rcParams['figure.figsize'] = [20,4]
# Make a scatterplot with distance from trench as x and depth as y:
plt.plot(bucaramanga["trench_dist_km_buc"], -bucaramanga["depth"], "ko", markersize=1)
# Add title and labels:
plt.title ('All earthquakes in Bucaramanga 01.01.1960 - 22.05.2018', color='b')
plt.xlabel('Distance from trench (km)', color='b')
plt.ylabel('Depth (km)', color='b')

# Display image:
plt.show()
```

## Cross-section looping for whole data

```python
# Import modules to import an image:
import matplotlib.cbook as cbook
import matplotlib.image as mpimg

# Set figure size:
plt.rcParams['figure.figsize'] = [20,8]
# Set path for saving images:
savepath="/Users/Nellimetiainen/Desktop/a/"

# For running value text box with month and year, assing names for values for 'months':
months = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}

# Select time range: first line years, second line months:
for year in range(1960, 2018):
    for month in range(1, 13):
        currentMonth = str(month)+"-"+str(year)
        selectedMonth = norm_sub[currentMonth:]
        
# Determine figure:
        fig = plt.figure()
# Add title for figure:
        fig.suptitle('Cross section view of earthquakes', fontsize=25, fontweight='bold')
        ax = fig.add_subplot(111)

# Set axis:
        distance,depth = (norm_sub["trench_dist_km_norm_sub"].values, -norm_sub["depth"].values)
        cur_dist,cur_depth = (selectedMonth["trench_dist_km_norm_sub"].values, -selectedMonth["depth"].values)

# Make scatter plot:
        plot = plt.scatter(cur_dist,cur_depth, c=selectedMonth["mag"], cmap='plasma_r')
# Add title:
        plt.title ('All earthquakes in norm_sub 01.01.1960 - 22.05.2018', color='b')
# Add labels:
        plt.xlabel('Distance from trench (km)', color='b')
        plt.ylabel('Depth (km)', color='b')
# Determine x & y axis:
        plt.axis ([0.0, distance.max(), depth.min(), 0.0])

# Determine colors:
        cmap = cm.get_cmap('plasma', 7)
# Create bounds for colorbar (min and max values for magnitude, and the number of separate values displayed):
        bounds = np.linspace(4, 10, 7)
# Set limits for colorbar min and max
        plot.set_clim(4.5, 9.5)
# 
        cbaxes = fig.add_axes([1400, -500, 50, 100])
# Create colorbar:
        cbar = plt.colorbar(cax = cbaxes, cmap=cmap, orientation="vertical", boundaries=bounds)  
# Add label:
        cbar.set_label('Magnitude', color='blue')

# Print month and year onto the figure in a colorbox:
        ax.text(1200,-500, str(year), color="white", fontsize=20, 
        name='Courier New',
        bbox={'facecolor':'blue', 'edgecolor':'none', 'alpha':0.7, 'pad':5})
        
# Add image:
# Set image size and location:
        axins = inset_axes(ax, width=1.3, height=0.9, loc=6)
# Read in image:
        with cbook.get_sample_data('/Users/Nellimetiainen/Desktop/south-america-map.gif') as image_file:
            image = plt.imread(image_file)
# Choose which part of the image to show:
            axins.imshow(image[200:300][:])
# Don't display axis:
            plt.axis("off")

# Save figure as picture:
        plotFile = str(year) + "-" + str(month) + "cross_section" + ".png"
        plt.savefig(savepath+plotFile)

# Save memory and avoid computer crashing by not opening all the figures in the loop:
        plt.close("all")
```
